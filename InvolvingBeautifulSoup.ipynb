```python
import bs4
import requests
```

```python
url="http://sistemas.opds.gba.gov.ar/intra/Profesionales/ConsultaWeb.php"
html_content = requests.get(url).text
```

```python
soup = bs4.BeautifulSoup(html_content, "lxml")
```

```python
print(soup.prettify())
```

```python
talbe = soup.find("tr", {"class":"FontNormal" })
```

```python
b = soup.find_all("td", {"class": "FontNormal"})
```

```python
options = soup.find_all("option")
for k, i in enumerate(options.get("value")):
    print(k,i)
```

```python
options_to_select = {"Titulo": [], "Partido": []}
for k,i in enumerate(options):
    if (k>5) & (k<390):
        options_to_select["Titulo"].append(i)
```

```python
tits = options_to_select["Titulo"]
```
